```python
import time
import json
import requests
import pandas as pd


# 获取数据
timenum = int(time.time()*1000)
url = f'https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5&callback=&_={timenum+1}'
res = requests.get(url)
data = json.loads(res.json()['data'])
data_set = []
china_data = data['areaTree'][0]['children']
for i in china_data:
    data_dict = {}
#     省份地区名称
    data_dict['province'] = i['name']
#     新增确诊
    data_dict['nowConfirm'] = i['total']['nowConfirm']
#     总确诊
    data_dict['confirm'] = i['total']['confirm']
#     疑似病例
    data_dict['suspect'] = i['total']['suspect']
#     死亡病例
    data_dict['dead'] = i['total']['dead']
#     治愈病例
    data_dict['heal'] = i['total']['heal']
#     死亡率
    data_dict['deadRate'] = i['total']['deadRate']
#     治愈率
    data_dict['healRate'] = i['total']['healRate']
    data_set.append(data_dict)
    

#  数据处理
df = pd.DataFrame(data_set)
# 数据保存
# df.to_csv(r'C:\Users\Shinelon\Desktop\国内疫情数据.csv') 
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map

# 数据可视化
china_map = (
    Map()
    .add('现有确诊', [list(i) for i in zip(df['province'].values.tolist(), df['nowConfirm'].values.tolist())])
    .set_global_opts(
        title_opts=opts.TitleOpts(title='各地区确诊人数',pos_top='48%',pos_left='65%'),
        visualmap_opts=opts.VisualMapOpts(max_=50, is_piecewise=True),
    )
)
china_map.render_notebook()
```
